```python
#Take the dataset and create dataframe
import pandas as pd
df = pd.read_html('/content/DailyWiseMarketArrivals.html')[-1]
df
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
#FB PROPHET - It accepts data only in form of 2 columns - 1.ds(relation of time),2.y(relative of price)
#Hence let us rename the date column as ds and Modal price column as y
df = df.rename(columns = {'Modal Price (Rs/q)':'y','Date':'ds'})
df
```

```python
df1 = df[['ds','y']]
df1
```

```python
df1.info()
```

```python
#The datatype of both ds and y columns are in string /object format
#Now let us convert the datatype of ds column into datetime format
#And let us convert the datatype of the y column into integer format
df1['ds'] = pd.to_datetime(df1['ds'])
df1['y'] = df1['y'].astype(int)
df1.info()
```

```python
#VISUALISATION
import matplotlib.pyplot as plt
plt.figure(figsize = (25,5))
plt.scatter(df1['ds'],df1['y'])
```

```python
#FBPROPHET
import prophet
op = prophet.Prophet()
op.fit(df1) #Here we are giving our data to the FB PROPHET LIBRARY
```

```python
forecast = op.make_future_dataframe(periods = 1000,freq = 'D')
forecast = op.predict(forecast)
```

```python
op.plot(forecast,xlabel = 'Date',ylabel = 'Potato Prices/q')
#The Balck dots - are the actual values of potato prices from 1st Jan 2005 to 17th sep 2022
#The light blue shade - It is the min and max value of potato prices , as predicted by our model
#The dark blue line - BEST FIT LINE
```
